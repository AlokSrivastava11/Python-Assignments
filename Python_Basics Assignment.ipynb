```python
# Question 2 : Explain the difference between List and Tuple in Python.

List

--You can change, add, or remove elements after creation. 
--Defined using square brackets []. 
--Consumes more memory. 
--Slower compared to Tuples. 
--Ex-my_list = [1, 2, 3] 

Tuples

--Once created, elements cannot be changed or modified. 
--Defined using parentheses (). 
--Consumes less memory (more efficient). 
--Faster, especially for large datasets. 
--Ex-my_tuple = (1, 2, 3) 
```

```python
#Question 3 : What is a function in Python? Why are functions useful?

--A function is a reusable block of code that performs a specific task. It only runs 
when it is called. You can pass data (parameters) into a function, and it can return data as a 
result. 
In Python, a function is defined using the def keyword. 

Functions are essential in programming for the following reasons: 

1. Code Reusability: You can write the code once and use it many times throughout 
your program without rewriting it. 

2. Modularity: They allow you to break down large, complex problems into smaller, 
manageable chunks. 

3. Better Readability: Code organized into functions is much easier to read and 
understand for other developers. 

4. Easy Maintenance: If you need to change how a specific task works, you only have 
to update the code inside the function once, rather than searching for every place 
that task is performed.
```

```python
# Question 8: Remove duplicates using a set
original_list = [10, 20, 10, 30, 40, 20, 50]


unique_list = list(set(original_list))

print("Original List:", original_list)
print("List after removing duplicates:", unique_list)
```

```python
# Question 9: Function to return the square of a number
def find_square(number):
    return number * number

n = int(input("Enter a number to find its square: "))
print(f"The square of {n} is {find_square(n)}")
```

```python
# Question 10: Count occurrences of a number in a list
numbers_list = [2, 3, 4, 2, 5, 2]
number_to_count = 2


result = numbers_list.count(number_to_count)

print(f"The list is: {numbers_list}")
print(f"The number {number_to_count} appears {result} times in the list.")
```
